# Aspectos teóricos y Algorítmos

## Glóbulos rojos

Los glóbulos rojos son los encargados de transportar el oxígeno en el cuerpo humano, la forma bicóncava facilita su deformación y así poder fluir por los capilares y en zonas de turbulencia.La sangre, siendo esta plasma+glóbulo rojo que es la que se testea en este experimento, se comporta como un fluído no Newtoniano debido al movimiento de los GR en el plasma. Los GR se describen como medios viscoelásticos que le permite tener facilidad a fluir. El comportamiento viscoelástico se puede describir con el modelo de Voigt-Kelvin, este modelo describe una combinación de las respuestas elásticas y viscosas de la siguiente manera

$F=\mu u + \eta \dot u$

donde F es la tensión aplicada, u la deformación y $\mu$ y $\eta$ los coeficientes de elesticidad y viscosidad respectivamente.

Además se debe tener en cuenta que los GR se juntan formando agregados o *rouleaux*, como pilas de monedas, y estas agrupaciones afectan su reología. Se especula que cuando el shear rate (tensión de corte) tiende a cero, prevalece la formación de agregados, hasta tener un gran agregado que se comporta como un sólido, este puede ser viscoelástico o viscoplástico. Cuando el shear rate (SR) incrementa los agregados tienden a romperse y la viscosidad de la sangre disminuye. Mientras más aumenta el SR, más evidente se hace la deformación de los GR, este comportamiento genera una disminución en la viscosidad. 

## Equipo de medición 
Los viscosímetros rotativos son instrumentos que dependen del hecho de que un cuerpo sólido rotando sumergido en un líquido está sometido a una fuerza de retardo debida al arrastre viscoso que es proporcional a la viscosidad del líquido. En el caso de esta investigación se tiene un viscosímetro cono-plato, que permite determinar la viscosidad de un fluido dado que se aplica una velocidad de corte conocida. El cono está enroscado a un eje que mide la "resistencia" del fluido a que el cono gira. Tiene varias escalas de velocidades medidas en rpm.

En la próxima linea se lee el archivo con pandas y se guarda del tipo DataFrame para luego poder trabajar con los datos

````python
#CE 1

import pandas as pd 


df=pd.read_csv('datos-viscosimetro.csv')
df
````

En la celda de ejecución 2 ~~,~~ se hace un filtro del DataFrame anterior para solo dejar los datos con SR=1,5rpm y así ver como se comportan los glóbulos rojos a bajas revolciones~~,~~. **E**n este caso se supone que todvía estan agregados ya que tienen bajo shear rate, como se mencionó en la sección *Glóbulos rojos* y podemos tener una idea de la agregación

````python
# CE 2
import numpy as np
import matplotlib.pyplot as plt

bajas_revoluciones=df[df['SR']==1.5][['Dias despues de irradiar','Muestra','n']]
m0_br=bajas_revoluciones[bajas_revoluciones['Muestra']=='M0'][['Dias despues de irradiar','n']]
m5_br=bajas_revoluciones[bajas_revoluciones['Muestra']=='M5'][['Dias despues de irradiar','n']]
m10_br=bajas_revoluciones[bajas_revoluciones['Muestra']=='M10'][['Dias despues de irradiar','n']]
m25_br=bajas_revoluciones[bajas_revoluciones['Muestra']=='M25'][['Dias despues de irradiar','n']]

plt.figure(figsize=(8,4))                                
plt.plot(m0_br["Dias despues de irradiar"],m0_br["n"],c= "r",label = "M0")
plt.plot(m5_br["Dias despues de irradiar"],m5_br["n"],c= "m",label = "M5")
plt.plot(m10_br["Dias despues de irradiar"],m10_br["n"],c= "g",label = "M10")
plt.plot(m25_br["Dias despues de irradiar"],m25_br["n"],c= "c",label = "M25")
plt.title("Evolución temporal de la viscosidad a bajas revoluciones")
plt.xlabel("Días despúes de irradiar")
plt.ylabel("Viscosidad[cp]")    
plt.ylim(-13,13) 
plt.legend()                          
  
plt.show()

 
````

En la Figura que se obtuvo en la CE 2 se observa que la viscosidad de la sangre entera medida a bajos SR presenta variaciones entre las distintas dosis. Aunque no hay una correlación entre las dosis de radiación y la alteración, se puede observar que luego del día 20 de almacenamiento las viscosidades de las muestras irradiadas presentan valores inferiores a los de la muestra sin irradiar en la mayoría de los casos.

Para poder entender el comportamiento en la siguiente celda se compara la viscosidad correspondiente al plasma respecto a sangre entera, que en este caso sería M0

````python
# CE 3
plasma_br=bajas_revoluciones[bajas_revoluciones['Muestra']=='Plasma'][['Dias despues de irradiar','n']]

plt.figure(figsize=(8,4))
plt.plot(m0_br["Dias despues de irradiar"],m0_br["n"],c= "r",label = "M0")
plt.plot(plasma_br["Dias despues de irradiar"],plasma_br["n"],c= "b",label = "Plasma")
plt.title("Comparación de la viscosidad del plasma con sangre entera")
plt.xlabel("Días despúes de irradiar")
plt.ylabel("Viscosidad[cp]")    
plt.ylim(0,15) 
plt.legend()                          
  
plt.show()
````

El plasma una variación de su viscosidad a medida que evolucionan los días, esto no debería suceder, ya que esta componente de la sangre no se sostuvo en tratamiento de irradiación y se mantuvo almacenado en la heladera con sus debidos conservantes. Los glóbulos rojos que se estudiaron se encontraban suspendidos en plasma al 40%. Esta diferencia podría ser la causa de la diferencia a bajas revoluciones que se ve en la celda anterior de código. 

Debido a este cambio en la viscosidad del plasma se calculó la viscosidad relativa, es decir *(viscosidad de la muestra-viscosidad del plasma)/viscosidad del plasma*, para eso se utilizó las herramietas de la libreria **Numpy**

````python
# CE 4
import numpy as np 

plasma_br
n_plasma=np.array(plasma_br['n'])
n_m0=np.array(m0_br['n'])
n_m5=np.array(m5_br['n'])
n_m10=np.array(m10_br['n'])
n_m25=np.array(m25_br['n'])
n_relativo_m0=(n_m0-n_plasma)/n_plasma
n_relativo_m5=(n_m5-n_plasma)/n_plasma
n_relativo_m10=(n_m10-n_plasma)/n_plasma
n_relativo_m25=(n_m25-n_plasma)/n_plasma

plt.figure(figsize=(8,4))
plt.plot(plasma_br["Dias despues de irradiar"],n_relativo_m0,c= "r",label = "M0")
plt.plot(plasma_br["Dias despues de irradiar"],n_relativo_m5,c= "m",label = "M5")
plt.plot(plasma_br["Dias despues de irradiar"],n_relativo_m10,c= "g",label = "M10")
plt.plot(plasma_br["Dias despues de irradiar"],n_relativo_m25,c= "c",label = "M25")

plt.title("Viscosidad relativa a bajas revoluciones")
plt.xlabel("Días despúes de irradiar")
plt.ylabel("Viscosidad[cp]")    
plt.ylim(-10,15) 
plt.legend()                          
  
plt.show()


````

En la figura de la celda posterior se observa que la viscosidad relativa no presenta diferencia significativa entre las distintas dosis, salvo el día 2. Se pueden ver dos picos anómalos para el día 23 y para el día 28, por lo cual se sugiere descartar estas mediciones.

Tampoco se observa variaciones significativas con el tiempo de almacenamiento. Es importante destacar que la no observación de variaciones significativas en esto ensayos puede estar relacionada con la poca sensibilidad de esta técnica para poder detectarlos dando el amplio rango del valor medio de la viscosidad obtenido.
