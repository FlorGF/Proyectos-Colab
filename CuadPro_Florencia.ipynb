#Proyectos computacionales (cuaderno 5)

**Calculos de cambio de entropía.**

 Sabemos que $\Delta S=\int^2_1 dq_{rev}/T$ donde T es la temperatura reversible. Como S es una función de estado solo depende de los estados final e inicial. Por lo tanto la fórmula utilizada antes para calcular la entropía de procesos ireversbles, buscando un proceso reversible que tenga el mismo estado inicial y final.

*Ejemplo*

 La capacidad calorífica específica a presión constante del agua es casi constante en el intervalo 15°C a 75°C su valor es 1.00cal/(g°K) a 1atm. Determinar $\Delta S$ cuando 100g de agua se calientan de 25 a 50°C a p=1atm

*Resolución* 

Primer tenemos que $dq=C_p dT$ y $C_p=c_p m=100cal/K$

Por lo tanto 
$\Delta S = \int_{T_1}^{T_2} \frac{Cp}{T} dT$ 




```python
from scipy import integrate
import numpy as np
x = np.linspace(298,323,1000,endpoint=True) #las temperaturas en kelvin
f=100/x 
Int=integrate.simps(f,x) 
print("El valor de la diferencia de entropía es:",Int,"cal/k")
```

```python
# Ejercicio 6.2
# cantidad de pacientes 20, probabilidad de seguir vivo 0.25
# como la probabilidad es vivir o morir, es decir un suceso o el otro, es una distribución nominal 
from math import factorial
import matplotlib . pyplot as plt
n=20 
p=0.25 
print("Valor esperado",n*p)
print("Varianza",n*p*(1-p))
#defino las conmutaciones
def C(a,r):
    return factorial(a) // factorial(r) // factorial(a-r)

#defino los valores del histograma
p_x=[]
for k in range(20):
  p_x.append(C(n,k)*p**k*(1-p)**(n-k))

print(p_x)
#grafico el histograma
plt.bar(range(20),p_x)                   # Histograma de 10 bins
plt.title("Histograma de probabilidades de vida para {} pacientes ".format(n))
plt.show()
```

```python
# ejercicios 9.1 y 9.2 
#cantidad de personas 500, mu=70 y sigma=10
#cuantas personas entre 60 y 80
#personas que superan 95
#intervalo de 95%

import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt

G=ss.norm(70,10)
x=np.arange(30,100,1)

cant1=int((G.cdf(80)-G.cdf(60))*500)

cant2=int((1-G.cdf(95))*500)



print("La cantidad de personas que esta entre 60kg y 80kg es: ",cant1)
print(cant2," Superan los 95kg")
```

Acordate que la CDF(X) es básicamente una integral desde infinito negativo (o el límite inferior de la distribución) hasta X. Para calcular la cantidd de personas entre 60 y 80 tenes que buscar el área entre 60 y 80. Eso te va a dar la "masa" de probabilidad contenida en esos límites. Si a ese valor lo multiplicas por la cantidad de gente total, te va a dar la cantidad de gente en ese rango. Lo mismo para calcular la cantidad de gente con peso>95: necesitás el área mś allá de 95 (acordate que el area total de una distribución de probabilidad es siempre 1)
